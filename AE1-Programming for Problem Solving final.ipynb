# Requirements (a)

```python
import csv
from tabulate import tabulate

# Function to load data from a CSV file
def csv_load_data(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            data.append(row)
    return data

# Function to retrieve device information based on OEM ID
def device_info_retrieve_by_oem_id(data, oem_id):
    result = []
    for device in data:
        if device['oem_id'] == oem_id:
            result.append({
                'Model Name': device['model'],
                'Manufacturer': device['manufacturer'],
                'Weight (grams)': device['weight_gram'],
                'Price': device['price'],
                'Price Currency': device['price_currency']
            })
    return result

# Function to retrieve device information based on code name
def device_info_retrieve_by_code_name(data, code_name):
    result = []
    for device in data:
        if device['codename'] == code_name:
            result.append({
                'Brand': device['brand'],
                'Model Name': device['model'],
                'RAM Capacity': device['ram_capacity'],
                'Market Regions': device['market_regions'],
                'Date Added': device['info_added_date']
            })
    return result

# Function to retrieve device information based on RAM capacity
def device_info_retrieve_by_ram_capacity(data, ram_capacity):
    result = []
    for device in data:
        if device['ram_capacity'] == str(ram_capacity):
            result.append({
                'OEM ID': device['oem_id'],
                'Release Date': device['released_date'],
                'Announcement Date': device['announced_date'],
                'Dimensions': device['dimensions'],
                'Device Category': device['device_category']
            })
    return result

# Function to retrieve custom device information based on specific conditions
def custom_info_retrieve(data):
    result = []
    for device in data:
        if float(device['weight_gram']) > 180 and device['device_category'] == 'Smartphone':
            result.append({
                'Model Name': device['model'],
                'Release Date': device['released_date'],
                'Price': device['price']
            })
    return result

# Function to print information using the tabulate library
def print_info(data, title):
    if data:
        headers = data[0].keys()
        rows = [list(row.values()) for row in data]
        print(f"\n{title}")
        print(tabulate(rows, headers=headers, tablefmt="fancy_grid"))
    else:
        print(f"\n{title}: No results found.")

# Main function to execute the tasks
def main():
    file_path = input("Enter the path to the CSV file: ")
    csv_data = csv_load_data(file_path)

    # Task a1: Function for the model name, manufacturer, weight, price, 
    # and price currency for the device(s) based on the oem_id.
    oem_id_to_retrieve = input("Enter the oem_id to retrieve (e.g., MZB0D87EU): ")
    result_a1 = device_info_retrieve_by_oem_id(csv_data, oem_id_to_retrieve)
    print_info(result_a1, "Task a1 Result")

    # Task a2:  Function for the brand, model name, RAM capacity, 
    # market regions, and the date when the data was added for the device(s) associated with a specified code name.
    code_name_to_retrieve = input("Enter the code name to retrieve (e.g., Xiaomi Ruby): ")
    result_a2 = device_info_retrieve_by_code_name(csv_data, code_name_to_retrieve)
    print_info(result_a2, "Task a2 Result")

    # Task a3: Function for the oem_id, release date, announcement date, dimensions, and 
    # device category of the device(s) based on a specified RAM capacity.
    ram_capacity_to_retrieve = int(input("Enter the RAM capacity to retrieve (e.g., 8): "))
    result_a3 = device_info_retrieve_by_ram_capacity(csv_data, ram_capacity_to_retrieve)
    print_info(result_a3, "Task a3 Result")

    # Task a4: Function for data from chosen columns and apply a specific condition that 
    # relates to an individual device.
    result_a4 = custom_info_retrieve(csv_data)
    print_info(result_a4, "Task a4 Result")

if __name__ == "__main__":
    main()
```

# Requirements (b)

```python
import pandas as pd
from tabulate import tabulate  

# Defined function to load data from a CSV file and handle file not found exception.
def load_csv_data(file_path):
    try:
        data = pd.read_csv(file_path)
        return data
    except FileNotFoundError:
        print("File not found. Please verify the file path.")
        return None

# Task b1: Identified the top 5 regions where a specific brand of devices was sold.
def top_regions_of_brand(brand_name, data):
    # Filter data for the specified brand and calculate the top regions.
    brand_data = data[data['brand'] == brand_name]
    top_regions = brand_data['market_regions'].value_counts().head(5)
    return top_regions

# Task b2: Analyze the average price of devices within a specific brand, all in the same currency.
def average_price_of_brand(brand_name, data):
    # Filter data for the specified brand and calculate the average price.
    brand_data = data[data['brand'] == brand_name]
    average_price = brand_data['price'].mean()
    return average_price

# Task b3: Analyze the average mass for each manufacturer and display the list of average mass for all manufacturers.
def average_mass_by_manufacturer(data):
    # Calculate the average mass grouped by manufacturer.
    average_mass = data.groupby('manufacturer')['weight_gram'].mean()
    return average_mass

# Task b4: Analyze the data to derive meaningful insights based on the unique selection.
def unique_insights_analysis_of_dataset(brand_name, data):
    # Filter data for the specified brand and perform various analyses.
    brand_data = data[data['brand'] == brand_name]

    # Calculate value counts for specific columns and other relevant details.
    column_counts = brand_data['device_category'].value_counts()
    feature1_counts = brand_data['ram_capacity'].value_counts()
    feature2_counts = brand_data['ram_type'].value_counts()
    feature3_counts = brand_data['manufacturer'].value_counts()

    # Calculate average price, average mass by manufacturer, and top regions.
    average_price = brand_data['price'].mean()
    average_mass = brand_data.groupby('manufacturer')['weight_gram'].mean()
    top_regions = brand_data['market_regions'].value_counts().head(5)

    # Create a dictionary to organize the insights.
    insights_dict = {
        "Brand": brand_name,
        "Top 5 Regions": top_regions,
        "Average Price": average_price,
        "Average Mass by Manufacturer": average_mass,
        "Model Value Counts": column_counts,
        "Ram Capacity Value Counts": feature1_counts,
        "Ram Type Value Counts": feature2_counts,
        "Manufacturer Value Counts": feature3_counts
    }

    return insights_dict

# Main function to execute tasks based on user input.
def main():
    # Prompt user for the dataset file path.
    file_path = input("Enter the dataset file path: ")
    
    # Load data from the specified file path.
    data = load_csv_data(file_path)
    if data is None:
        return

    # Prompt user for the brand name.
    brand_name = input("Enter the brand name: ")

    # Call the function to obtain unique insights for the specified brand.
    brand_insights = unique_insights_analysis_of_dataset(brand_name, data)

    # Display the obtained insights.
    for key, value in brand_insights.items():
        if isinstance(value, pd.Series):
            print(f"{key}:\n{tabulate(pd.DataFrame({key: value}))}")
        else:
            print(f"{key}: {value}")

if __name__ == "__main__":
    main()
```

# Requirements (c)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to load data from a CSV file and handle file not found exception.
def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        return data
    except FileNotFoundError:
        print("File not found. Please verify the file path.")
        return None

# Task c1: Generate a pie chart depicting the proportion of RAM types in devices.
def ram_type_proportion_plot(data):
    # Calculate the counts of each RAM type.
    ram_type_counts = data['ram_type'].value_counts()

    # Create and display a pie chart.
    plt.figure(figsize=(8, 8))
    plt.pie(ram_type_counts, labels=ram_type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#99ff99', '#ffcc99'])
    plt.title('Proportion of RAM Types for Devices')
    plt.show()

# Task c2: Generate a bar chart comparing the number of devices for each USB connector type.
def usb_connector_comparison_plot(data):
    # Calculate the counts for each USB connector type.
    usb_connector_counts = data['usb_connector'].value_counts()

    # Create and display a bar chart.
    plt.figure(figsize=(10, 6))
    usb_connector_counts.plot(kind='bar', color=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'])
    plt.title('Number of Devices for Each USB Connector Type')
    plt.xlabel('USB Connector Type')
    plt.ylabel('Number of Devices')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Task c3: Generate separate line charts illustrating monthly average price trends for devices released each year from 
# 2020 to 2023.
def average_price_trends_plot(data):
    # Convert 'released_date' to datetime and extract the year.
    data['released_date'] = pd.to_datetime(data['released_date'], format='%d-%m-%y', errors='coerce')
    data['year'] = data['released_date'].dt.year

    # Define the years of interest.
    years = list(range(2020, 2024))

    # Create line charts for each year's monthly average price trends.
    for i, year in enumerate(years):
        year_data = data[data['year'] == year]
        monthly_average_price = year_data.groupby(year_data['released_date'].dt.month)['price'].mean()

        colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
        plt.figure(figsize=(10, 6))
        plt.plot(monthly_average_price.index, monthly_average_price.values, marker='o', linestyle='-', color=colors[i])
        plt.title(f'Monthly Average Price Trends in {year} (in GBP)')
        plt.xlabel('Month')
        plt.ylabel('Average Price (GBP)')
        plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
        plt.show()

# Task c4: Generate scatter plots to visualize relationships between selected device features and prices.
def device_features_visualization_plot(data):
    # Define selected features for visualization.
    selected_features = ['pixel_density', 'device_category', 'platform']

    # Create scatter plots for each selected feature against device prices.
    for feature in selected_features:
        plt.figure(figsize=(10, 6))
        plt.scatter(data[feature], data['price'], alpha=0.5, color='#FF6347')
        plt.title(f'{feature} vs. Price')
        plt.xlabel(feature)
        plt.ylabel('Price')
        plt.show()

# Main function to execute tasks based on user input.
def main():
    # Prompt user for the dataset file path.
    file_path = input("Enter the dataset file path: ")

    # Load data from the specified file path.
    data = load_data(file_path)
    if data is None:
        return

    # Task c1: Generate the RAM type proportion chart.
    ram_type_proportion_plot(data)

    # Task c2: Generate the USB connector comparison chart.
    usb_connector_comparison_plot(data)

    # Task c3: Generate the monthly average price trends charts.
    average_price_trends_plot(data)

    # Task c4: Generate scatter plots for device features and prices.
    device_features_visualization_plot(data)

if __name__ == "__main__":
    main()
```
